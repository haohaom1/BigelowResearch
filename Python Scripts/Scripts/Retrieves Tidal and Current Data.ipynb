# Cleans the current data

```python
csvs = ['Tidal_Current Data/CAB1409_prediction_Timeseries_{}.csv'.format(d) for d in 
        ['2017-03-17', '2017-05-02', '2017-12-19', '2018-03-28', '2018-05-07']]
flood_dir = [354, 354, 354, 354, 354]
ebb_dir = [179, 179, 179, 179, 179]
```

```python
df0 = pd.read_csv(csvs[3], nrows=240)
# adds seconds
df0['Date_Time (GMT)'] = df0['Date_Time (GMT)'].apply(lambda x: x[:-3] + ':00' + x[-3:])
```

```python
df0['direction (degrees)'] = np.where(df0[' Speed (cm/sec)'] > 0, 354, 179)
```

```python
df0
```

```python
for csv, flood, ebb in zip(csvs, flood_dir, ebb_dir):
    df = pd.read_csv(csv, nrows=240)
    df.rename(index=str, columns={' Speed (cm/sec)': 'Speed (cm/sec)'}, inplace=True)
    df['Date_Time (GMT)'] = df['Date_Time (GMT)'].apply(lambda x: x[:-3] + ':00' + x[-3:]) # adds seconds
    df['Direction (degrees relative to North)'] = np.where(df['Speed (cm/sec)'] > 0, flood, ebb)
    df.to_csv(csv[:-4] + '_cleaned.csv', index=False)
```
